```python
#iris dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
```

```python
# load dataset into Pandas DataFrame
df = pd.read_csv(url, names=['sepal length','sepal width','petal length','petal width','target'])
df.shape
```

```python
from sklearn.preprocessing import StandardScaler
features = ['sepal length', 'sepal width', 'petal length', 'petal width']
features
```

```python
# Separating out the features
x = df.loc[:, features].values
# Separating out the target
y = df.loc[:,['target']].values
x[0:5],y[0:5]
```

```python
# Standardizing the features
x = StandardScaler().fit_transform(x)
x[0:5]
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
```

```python
principalComponents = pca.fit_transform(x)
principalComponents
```

```python
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
principalDf
```

```python
#PCA and keeping the top two principal components
finalDf = pd.concat([principalDf, df[['target']]], axis = 1)
finalDf
```

```python
# Visualize 2D Projection
#This section is just plotting two-dimensional data. Notice on the graph below that the classes seem well separated from each other.

fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)

targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['target'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
plt.show();
```

```python
#Explained Variance
#The explained variance tells you how much information (variance) can be attributed to each of the principal components.
#This is important because while you can convert four-dimensional space to a two-dimensional space, you lose some of the 
#variance (information) when you do this. By using the attribute explained_variance_ratio_, 
#you can see that the first principal component contains 72.77 percent of the variance, and 
#the second principal component contains 23.03 percent of the variance. Together, the two components contain 95.80 percent 
#of the information.

pca.explained_variance_ratio_
```

```python
# mtcars
from pydataset import data
mtcars = data('mtcars')
mtcars.head()
```

```python
X = mtcars[['mpg','cyl','hp','gear','wt','carb']].values
y = mtcars['am'].values
X[0:5], y[0:5]
```

```python
PCA
```

```python
X = StandardScaler().fit_transform(X)
X[0:5]
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(X)
principalComponents[0:5]
```

```python
principalMT = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
principalMT[0:5]
```

```python
print(principalMT.shape, y.reshape(-1,1).shape)
print(type(principalMT), type(pd.Series(y).reset_index()))
ymt = pd.Series(y)
ymt
```

```python
#PCA and keeping the top two principal components
finalMT = pd.concat([principalMT, ymt], axis = 1)
finalMT.columns = ['PC1','PC2','Tx']
finalMT
```

```python
pca.explained_variance_ratio_
```

```python
y
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('PC1', fontsize = 15)
ax.set_ylabel('PC2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)

targets = [0,1]
colors = ['r', 'g']
for target, color in zip(y,colors):
    indicesToKeep = finalMT['Tx'] == target
    ax.scatter(finalMT.loc[indicesToKeep, 'PC1'], finalMT.loc[indicesToKeep, 'PC2'] , c = color, s = 50)
ax.legend(targets)
ax.grid()
plt.show();
```
