# subplots
-  

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pydataset import data
import seaborn as sns
```

```python
mtcars = data('mtcars')
```

```python
mtcars.plot(kind = 'density', subplots = True, layout=(3,4), sharex=False, figsize=(10,8))
plt.show()
```

```python
mtcars.plot(kind='box', subplots=True, layout = (3,4), sharex=False, sharey=False, figsize=(10,8))
plt.show()
```
