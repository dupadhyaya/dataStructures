# Histograms in python

```python
import numpy as np
import pandas as pd
from pydataset import data
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
#hist1
salesA = np.random.randint(10,100,1000)
salesB = np.random.randint(70,120,1000)
print(salesA[0:5], salesB[0:5])
```

```python
# Plotting a basic histogram
plt.hist(salesA, bins=10, color='skyblue', edgecolor='black')
# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')
 
# Display the plot
plt.show()
```

```python
# Plotting a basic histogram
sns.histplot(salesA, bins=10, color='skyblue', edgecolor='black', kde=True)
plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(8,6))
axs[0].hist(salesA, bins=10, color='yellow', edgecolor='black', label='Sales A')
axs[1].hist(salesB, bins=5, color='green', edgecolor='blue', label='Sales B', density=True) #density
axs[0].set_title('Sales A')
axs[1].set_title('Sales B')
fig.suptitle('Histogram showing Sales of Coy A & B')
for a in axs:
    a.set_xlabel('Sale Interval')
    a.set_ylabel('Count at Sales Intervals')
fig.legend()
plt.tight_layout()
plt.show()
```

```python
plt.hist([salesA, salesB], bins=10, stacked=True, color=['yellow','cyan'], edgecolor='black')
plt.legend(['SalesA', 'SalesB'] )
plt.xlabel('Sales Interval')
plt.ylabel('Count at these intervals')
plt.title('Stacked Histgram')
plt.show()
```

```python
# specify bins
plt.hist(salesA, bins=[0,40,50,60,70,80,100], color='cyan', edgecolor='black')
plt.show()
```

```python
#label values
values, bins, bars = plt.hist(salesA, edgecolor='black')
plt.xlabel('Sales Interval', loc='right')
plt.text(x=20, y=20, s='Sales Interval in X Axis', bbox=dict(facecolor='red'), rotation=45)
plt.ylabel('Count of Days with Sales Interval', loc='bottom')
plt.title('Histogram with Count')
plt.bar_label(bars, fontsize=15, color='red')
plt.margins(x=.01, y=.1)
plt.show();
```
