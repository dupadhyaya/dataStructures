```python
# Bar plots
- Stacked Bar
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pydataset import data
```

```python
tips = data('tips')
tips.head()
```

```python
#gender - smoker wise tips
```

```python
df = tips.groupby('sex')['smoker'].value_counts(normalize=True).mul(100).round(2)
df
```

```python
df = df.unstack()
df
#: use unstack( ) method so that the sex labels will be presented in index and 
#smoker status in columns and percentage values in cells.
```

```python
fig, ax = plt.subplots()
df.plot(kind='bar', stacked=True, ax=ax, edgecolor='black')
plt.show();
```

```python
fig, ax = plt.subplots()
df.plot.bar(stacked=True, ax=ax, edgecolor='black')
plt.show();
```

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.bar(df.index, df['No'], label='No')
ax.bar(df.index, df['Yes'], bottom =df.No, label='Yes')
plt.show();
```

```python
ax.containers
```

```python
# Adding bar labels
fig, ax = plt.subplots(figsize=(10,8))
ax.bar(df.index, df['No'], label='No')
ax.bar(df.index, df['Yes'], bottom =df.No, label='Yes')
for c in ax.containers:
    labels = [str(round(v.get_height(), 2)) + "%" if v.get_height() > 0 else '' for v in c]
    ax.bar_label(c, label_type='center',   labels = labels,size = 14) 
    # add a container object "c" as first argument
# Removing spines
for s in ["top", "right"]:
    ax.spines[s].set_visible(False)
# Adding tick and axes labels
ax.tick_params(labelsize = 14, labelrotation = 0)
ax.set_ylabel("Percentage", size = 14)
ax.set_xlabel("Sex", size = 14)
# Customising legend
ax.legend(labels = ["no", "yes"],fontsize = 12, title = "Smoker", title_fontsize = 18)
# Fixing legend position
ax.legend_.set_bbox_to_anchor([0.55, 0.7])
```
