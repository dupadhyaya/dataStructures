```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from pydataset import data
```

```python
mtcars = data('mtcars')
mtcars.head()
```

```python
# pie1
y = np.array([35,25,25, 15])
plt.pie(y)
plt.show()
```

```python
# pie2
y = np.array([35,25,25, 15])
depts = ['IT','HR','Marketing','Engg']
plt.pie(y, labels=depts)
plt.show()
```

```python
# pie3
y = np.array([35,25,25, 15])
depts = ['IT','HR','Marketing','Engg']
plt.pie(y, labels=depts, startangle=0)  #default 0
plt.show()
```

```python
# pie4
# explode = one of wedges to stand out; shadow
y = np.array([35,25,25, 15])
depts = ['IT','HR','Marketing','Engg']
myexplode =[.2,0,0,0]
plt.pie(y, labels=depts, explode=myexplode, shadow=True) 
plt.show();
#myshadow={'ox': -0.04, 'edgecolor': 'none', 'shade': 0.9}
```

```python
# pie4 : colors & legend
# colors 'r' - Red, 'g' - Green, 'b' - Blue, 'c' - Cyan, 'm' - Magenta, 'y' - Yellow, 'k' - Black, 'w' - White
y = np.array([35,25,25, 15])
depts = ['IT','HR','Marketing','Engg']
mycolors =['r','g','blue','black']
plt.pie(y, labels=depts, colors=mycolors) 
plt.legend(title='Departments', loc='center right')
plt.show()
```

```python
# auto label slices
y = np.array([35,25,25, 15])
depts = ['IT','HR','Marketing','Engg']
mycolors =['r','g','blue','y']
plt.pie(y, labels=depts, colors=mycolors, autopct='%1.1f%%')
plt.show();
```

```python
# hatch patterns  "*+-./OX\ox|"
myhatch = ['**+', 'oX', 'Xo', '.\\.']
plt.pie(y, labels=depts, hatch=myhatch)
plt.show();
```

```python
#sway label & autopct
plt.pie(y, labels=depts, autopct='%1.1f%%', pctdistance=1.25, labeldistance=.6)
plt.show();
```

```python
#size
plt.pie(y, labels=depts, autopct='%.0f%%', textprops={'size': 'smaller'}, radius=1.5)
plt.show();
```

```python
# pandas
gearCount1 = mtcars.groupby('gear',as_index=False).agg(count=('am','size'))
gearCount2 = mtcars.groupby('gear').agg(count=('am','size'))
```

```python
gearCount1
```

```python
list(gearCount1['gear'].astype('string'))
```

```python
#mpl
plt.pie(gearCount1['count'], labels=gearCount1['gear'], autopct='%1.1f%%')
plt.show()
```

```python
#pd
gearCount1.plot.pie(y='count', autopct='%1.1f%%', labels=['G3','G4','G5'])
plt.show()
```

```python
gearCount2
```

```python
gearCount2.plot.pie(y='count', subplots=True, autopct='%1.1f%%')
plt.show()
```

```python
gearCount2.plot.pie(y='count', labels=gearCount2['count'])
plt.show()
```
