```python
titanic2.stb.subtotal()
#rying to add a subtotal to grouped pandas data is not easy. sidetable adds a subtotal() function that makes adds a subtotal at one or more levels of a DataFrame.
```

```python
titanic2.groupby(['sex', 'class', 'embark_town']).agg({'fare': ['sum']}).stb.subtotal()
```

```python
summary_table = titanic2.groupby(['sex', 'class', 'embark_town']).agg({'fare': ['sum']})
summary_table.stb.subtotal(sub_level=[1, 2])
```

```python
summary_table.stb.subtotal(sub_level=[1])
```

```python
#flatten
#When grouping and pivoting data, you can end up with a DataFrame that has a multiindex. Often times, you want a simple flat representation of the data.
#For example, we can build a table using a groupby() plus unstack() that looks like this:

titanic2.groupby(['embark_town', 'class', 'sex']).agg({'fare': ['sum'], 'age': ['mean']}).unstack()
```

```python
titanic2.groupby(['embark_town', 'class', 'sex']).agg({'fare': ['sum'], 'age': ['mean']}).unstack().stb.flatten()
```

- flatten will also take additional arguments:
- Add a custom separator using the sep argument - stb.flatten(sep='|')
- Control whether or not to reset the index using reset argument - stb.flatten(reset=False)
- Reorganize the output levels using levels argument levels=2
- levels can also take a list of valid levels if you want to reorganize the display levels=[0,2]

```python
fares = titanic2.groupby(['embark_town', 'class', 'sex']).agg({'fare': ['sum'], 'age': ['mean']}).unstack()
fares.stb.flatten(sep='_', reset=False, levels=[0,2])
```

# prettyprint
- This function interprets the magnitude of your numeric results and returns a nicely formatted version of all the numbers. This can be used on a full DataFrame or during your analysis of aggregated data.
- For instance, if you are summarizing data, you may get something that looks like this:


```python
titanic2.groupby(['pclass', 'sex']).agg({'fare': 'sum'})
```

```python
titanic2.groupby(['pclass', 'sex']).agg({'fare': 'sum'}).div(titanic2['fare'].sum()).stb.pretty()
```

```python
titanic2.groupby(['pclass', 'sex']).agg({'fare': 'sum'}).div(titanic2['fare'].sum()).stb.pretty(precision=0, caption="Fare Percentage")
```
