```python
airp = data('AirPassengers')
airp.head()
```

```python
from datetime import datetime, timedelta
airp['date']= airp['time'].apply(lambda x : datetime(int(x), 1, 1) + timedelta(days = (x % 1) * 365+1))
airp.head()
```

```python
airp['monYr'] = airp['date'].dt.strftime('%b-%Y')
airp.head()
```
