```python
#libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Types of clustering

```python
# data creation
from sklearn.datasets import make_blobs
```

```python
# 3 clusters with 150 randomly generated data points and 2 features
X, y = make_blobs(n_samples=150, n_features=2, centers=3, cluster_std=0.5, shuffle=True, random_state=0)
X, y
# X is an array of 2 cols x 150 rows, y - array of 1 col x 150 rows showing which group X belong
```

```python
plt.scatter( X[:,0], X[:,1], marker='x', s=50)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid()
plt.tight_layout()
plt.show();
```

```python
# in reallife we don't the groups they belong to (ie y)
# so find the no of groups through kmeans
```

```python
from sklearn.datasets import make_classification
from sklearn.cluster import KMeans
```

```python
# generate new data 
X, y = make_classification( n_samples = 1000, n_features=20, random_state=42)
X[0:5,0:4]   #sample data
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=5)
kmeans
```

```python
#fit the model into existing data
cluster_labels = kmeans.fit_predict(X)
```

```python
# Create a scatter plot for the first two features
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Data with Clusters')
plt.colorbar(label='Cluster')
plt.show();
```
